```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error,mean_absolute_error,r2_score
```

```python
np.random.seed(0)
hours_studied = np.random.uniform(0, 10, 100)  
exam_scores = 50 + 5 * hours_studied + np.random.normal(0, 2, 100) 
```

```python
data = pd.DataFrame({'Hours_Studied': hours_studied, 'Exam_Scores': exam_scores})
```

```python
data.isnull().sum()
```

```python
data.duplicated().sum()
```

# Data Spliting

```python
x = data[['Hours_Studied']]  # here is a one feature so it called simple linear
y = data['Exam_Scores']  
```

```python
X_train,X_test,Y_train,Y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

# Linear Regression Modeling

```python
model = LinearRegression()
model.fit(X_train,Y_train)
```

# Model Evolution

```python
Y_pred = model.predict(X_test)
Y_pred
```

# Model Accuracy

```python
MSE = mean_squared_error(Y_test,Y_pred)
MAE = mean_absolute_error(Y_test,Y_pred)
R2 = r2_score(Y_test,Y_pred)
```

```python
print(f'mean_absolute_error : {MAE:.2f}')
print(f'mean_squared_error : {MSE:.2f}')
print(f'r2_score : {R2:.2f}')
```

# Visulization

```python
plt.scatter(X_test,Y_test,color = 'b',label='Simple Linear Regression')
plt.plot(X_test,Y_pred,color='r',linewidth=2)
plt.xlabel('Hours Studied')
plt.ylabel('Exam Scores')
plt.legend()
plt.title('Simple Linear Regression: Hours Studied vs. Exam Scores')
plt.legend()
plt.show()
```
